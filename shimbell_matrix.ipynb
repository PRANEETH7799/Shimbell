### Shimbell matrix verification function

```python
def fun(mat):
  non_diag = np.ones(shape=mat.shape, dtype=bool) ^ np.identity(len(mat)).astype(bool) # https://stackoverflow.com/a/35746922
  bool_mat = np.array(mat, dtype=bool)
  return np.array_equal(bool_mat,non_diag)
```

## Shimbell matrix calculation function

```python
def dyad_calculation(c):
  n = 1
  k = np.identity(len(c))
  d = np.zeros([len(c),len(c)])
  while not fun(d):
    k = np.dot(c,k)
    print("The matrix multiplication is \n", k)
    for i in range(0,len(c)):
      for j in range(0,len(c)):
        if k[i][j]!=0 and d[i][j]==0:
          d[i][j] = n
      d[i][i] = 0
      
    n = n + 1
    print("The D value is \n",d)
    print(fun(d))
  print("The dyad value is \n",d)
  print("The shortest paths from each vertices to other vertices", np.sum(d,axis=0))
```

```python
dyad_calculation(c)
```
